## 🎯 Öğrenme Hedefleri (Learning Objectives)

- Python'da küçük bir ağaç ve graf yapısı kurabilmek
- Genişlik öncelikli arama / Breadth-First Search (BFS) ile derinlik öncelikli arama / Depth-First Search (DFS) algoritmalarını uygulayabilmek
- Hangi durumda BFS, hangi durumda DFS seçileceğini gerekçelendirmek
- Karmaşıklığı **O(V + E)** olarak ifade etmek

## ⚙️ Ön Koşullar (Prerequisites)
Listeler, sözlükler (dict), kuyruk (queue), yığıt (stack), özyineleme (recursion).


````python
#Hatırlatma
## Python'da list ile stack
stack = []

stack.append("A")   # push
stack.append("B")
stack.append("C")

print(stack.pop())  # C  ← son giren, ilk çıkar
print(stack.pop())  # B
print(stack)        # ['A']


## Python'da Queue
from collections import deque

queue = deque()

queue.append("A")    # enqueue
queue.append("B")
queue.append("C")

print(queue.popleft())  # A  ← ilk giren, ilk çıkar
print(queue.popleft())  # B
print(queue)            # deque(['C'])
````

## ⚡ Hızlı Özet (Quick Summary)

- **Ağaç (Tree):** Düğümlerin (node) hiyerarşik bağlandığı, **döngüsüz**, **tek köklü** yapı.
- **Graf (Graph):** Düğümler (vertex) ve aralarındaki kenarlar (edge). **Döngü olabilir**, kök yoktur.
- **BFS (Breadth-First Search):** Kuyruk (queue) ile çalışır. Önce yakını tarar.
- **DFS (Depth-First Search):** Yığıt (stack) veya özyineleme ile çalışır. Önce derine iner.

> "Tüm ağaçlar graftır, ama tüm graflar ağaç değildir." — *Wengrow, Bölüm 18*

## 🤔 Neden Önemli? (Why It Matters)
Sosyal ağlar, navigasyon, web tarayıcıları, dosya sistemleri, bilgisayar ağları, oyun yapay zekâsı — hepsinin altında **ilişkili veri** vardır. Bu ilişkileri doğru ifade eden iki yapı: **ağaç ve graf**. Onları gezmek için kullanılan iki temel yöntem: **BFS ve DFS**. Bu iki algoritmayı bilmek, yüzlerce gerçek problemin çözümünün anahtarıdır.

### 🧠 Akılda Kalsın (What to Remember)
- Ağaç = döngüsüz, tek köklü graf
- BFS = kuyruk + yakını önce → **en kısa adım sayısı**
- DFS = stack/recursion + derine önce → **bir yere hızlıca in**
- Graf gezerken `visited` setini **mutlaka** tut, yoksa sonsuz döngü
- Hem BFS hem DFS karmaşıklığı: **O(V + E)**

## 🌳 Veri Yapısı 1: Ağaç (Tree)

### Temel Fikir (Mental Model)
Ağacı **tepeden çiziyoruz** — kök yukarıda, yapraklar aşağıda.

**Analoji:** Bir şirketin organizasyon şeması. CEO kök (root); her yönetici altındaki çalışanların ebeveyni; en alttaki çalışanlar yaprak (leaf).

```
        A          <- kök (root)
       / \
      B   C        <- A'nın çocukları (children)
     / \   \
    D   E   F      <- yaprak (leaves)
```

### Önemli Kurallar
1. **Tek bir kök** vardır.
2. Her düğümün (kök hariç) **bir** ebeveyni vardır.
3. **Döngü yoktur.**

### Operasyonlar
- Ekleme (insert), silme (delete), arama (search), gezinti (traversal)

### Artı / Eksi
| ✅ Avantaj | ❌ Dezavantaj |
|---|---|
| Hiyerarşik veriler için doğal (dosya sistemi, DOM, organigramlar) | Esnek ilişkiler için yetersiz (sosyal ağ → graf gerek) |
| Dengeli ikili arama ağacında arama O(log N) | Dengesizleşirse O(N)'e düşer |


````python
class TreeNode:
    """Basit ağaç düğümü: değer + çocuk listesi."""
    def __init__(self, value):
        self.value = value
        self.children = []

    def add_child(self, child_node):
        """Bir alt düğüm (çocuk) ekler."""
        self.children.append(child_node)


# Küçük örnek ağaç:
#         A
#        / \
#       B   C
#      / \   \
#     D   E   F

root = TreeNode("A")
b = TreeNode("B"); c = TreeNode("C")
d = TreeNode("D"); e = TreeNode("E"); f = TreeNode("F")

root.add_child(b); root.add_child(c)
b.add_child(d); b.add_child(e)
c.add_child(f)

# Hızlı testler (assert-based)
assert root.value == "A"
assert len(root.children) == 2
assert b.children[0].value == "D"
print("Ağaç kuruldu ✅")
print("  Kök:", root.value)
print("  B'nin çocukları:", [ch.value for ch in b.children])
print("  C'nin çocukları:", [ch.value for ch in c.children])
````

## 🕸️ Veri Yapısı 2: Graf (Graph)

### Temel Fikir
Graf, **düğümler (vertices)** ve aralarındaki **kenarlardan (edges)** oluşur. Ağaçtan farkı: **döngü olabilir** ve **tek bir kök yoktur**.

**Analoji:** Bir sosyal ağ. Ayşe, Berk'in arkadaşı; Berk, Defne'nin arkadaşı; Defne, Ayşe'nin arkadaşı → bir **döngü (cycle)** oluştu.

```
   Ayşe ─── Berk
    │ \      │
    │  \     │
   Defne ─── Furkan
              │
            Elif
```

### Ağaç vs Graf
| Özellik | Ağaç (Tree) | Graf (Graph) |
|---|---|---|
| Döngü (cycle) | ❌ Yok | ✅ Olabilir |
| Tek kök | ✅ Var | ❌ Yok |
| Tüm düğümler bağlı | ✅ Evet | ❌ Olmayabilir |

### Python'da Gösterim: Bitişik Liste (Adjacency List)
En basit yol: bir **dict**. Anahtar = düğüm, değer = komşuların listesi. Bu yapı arama için **O(1)** komşu erişimi sağlar (Wengrow, s.331).


````python
# Sosyal ağ örneği (Wengrow, Bölüm 18 — uyarlanmış)
graph = {
    "Ayşe":   ["Berk", "Defne", "Furkan"],
    "Berk":     ["Ayşe", "Ceyda", "Defne"],
    "Ceyda": ["Berk"],
    "Defne":   ["Ayşe", "Berk", "Furkan"],
    "Elif":   ["Furkan"],
    "Furkan":    ["Ayşe", "Defne", "Elif"],
}

# Hızlı testler
assert "Berk" in graph["Ayşe"]
assert "Ayşe" in graph["Berk"]      # iki yönlü (undirected)
assert len(graph["Defne"]) == 3
print("Graf kuruldu ✅")
print("  Ayşe'nin arkadaşları:", graph["Ayşe"])
print("  Toplam düğüm sayısı:", len(graph))
print("  Toplam kenar sayısı (yönsüz):", sum(len(v) for v in graph.values()) // 2)
````

## 🚀 Algoritma 1: Genişlik Öncelikli Arama — BFS (Breadth-First Search)

### Problem Tanımı
> "Bir başlangıç düğümünden hedef düğüme **en az kaç adımda** ulaşırım?"  
> Ya da: "Bir düğümün **doğrudan komşularını / yakın çevresini** önce gezeyim."

### Neden Kuyruk (Queue)?
BFS, bir düğümün **tüm komşularını** ziyaret etmeden bir alt seviyeye geçmemelidir. Yani **önce ekleneni önce işlemek** gerekir → **FIFO** → **kuyruk**.

### Algoritma (Wengrow, s.348)
1. Başlangıç düğümünü `visited` kümesine ekle ve kuyruğa koy.
2. Kuyruk boş değilken:
   - Kuyruktan bir düğüm çek (`current`).
   - `current`'in komşularını gez.
   - **Ziyaret edilmemiş** komşuları `visited`'e ekle ve kuyruğa koy.

### Adım Adım Çalışma — Ayşe'den BFS
```
Adım 0: queue=[Ayşe]                visited={Ayşe}
Adım 1: çek Ayşe → komşular: Berk, Defne, Furkan
        queue=[Berk, Defne, Furkan]     visited+={Berk,Defne,Furkan}
Adım 2: çek Berk → Ceyda yeni
        queue=[Defne, Furkan, Ceyda] visited+={Ceyda}
Adım 3: çek Defne → tümü ziyaret
        queue=[Furkan, Ceyda]
Adım 4: çek Furkan → Elif yeni
        queue=[Ceyda, Elif]       visited+={Elif}
Adım 5-6: Ceyda, Elif — yeni komşu yok
Sonuç: Ayşe → Berk → Defne → Furkan → Ceyda → Elif
```


````python
from collections import deque

def bfs(graph, start):
    """Graf üzerinde BFS gezintisi. Ziyaret sırasını döner."""
    visited = {start}
    queue = deque([start])
    order = []

    while queue:
        current = queue.popleft()           # FIFO — önce gireni çek
        order.append(current)
        for neighbor in graph[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return order


def bfs_shortest_path(graph, start, target):
    """Ağırlıksız grafta start -> target en kısa yolu döner (yoksa None)."""
    if start == target:
        return [start]
    visited = {start}
    queue = deque([(start, [start])])       # (düğüm, oraya kadar yol)
    while queue:
        current, path = queue.popleft()
        for neighbor in graph[current]:
            if neighbor not in visited:
                if neighbor == target:
                    return path + [neighbor]
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    return None


print("BFS sırası (Ayşe'den):", bfs(graph, "Ayşe"))
print("Ayşe → Ceyda en kısa yol:", bfs_shortest_path(graph, "Ayşe", "Ceyda"))
print("Ayşe → Elif   en kısa yol:", bfs_shortest_path(graph, "Ayşe", "Elif"))
````

## 🌊 Algoritma 2: Derinlik Öncelikli Arama — DFS (Depth-First Search)

### Problem Tanımı
> "Mümkün olduğunca **derine** ineyim; çıkmaza geldiğimde geri dönüp başka dalı denerim."

### Neden Yığıt (Stack) veya Özyineleme?
DFS, bir komşuya iner inmez **o komşunun komşularını** işlemek ister. Yani **en son gireni önce** işlemek gerekir → **LIFO** → **yığıt**. Özyineleme de doğal olarak çağrı yığıtını (call stack) kullandığı için, DFS özyinelemeyle de yazılabilir.

### Algoritma (Wengrow, s.339)
1. Mevcut düğümü `visited`'e ekle.
2. Komşularını gez:
   - Ziyaret edilmemiş her komşu için **özyinelemeli** olarak DFS çağır.

### Adım Adım Çalışma — Ayşe'den DFS (alfabetik sıra)
```
DFS(Ayşe)
  visited = [Ayşe]
  ↓ ilk komşu: Berk
  DFS(Berk)
    visited = [Ayşe, Berk]
    ↓ ilk komşu (Ayşe ✓): Ceyda
    DFS(Ceyda)  → yaprak, geri dön
    ↓ sonraki: Defne
    DFS(Defne)
      ↓ Furkan
      DFS(Furkan)
        ↓ Elif
        DFS(Elif)  → yaprak
Sonuç: Ayşe → Berk → Ceyda → Defne → Furkan → Elif
```


````python
def dfs(graph, start, visited=None, order=None):
    """Özyinelemeli DFS gezintisi."""
    if visited is None: visited = set()
    if order is None:   order = []
    visited.add(start)
    order.append(start)
    for neighbor in sorted(graph[start]):   # alfabetik sıra (deterministik test için)
        if neighbor not in visited:
            dfs(graph, neighbor, visited, order)
    return order


def dfs_iterative(graph, start):
    """Stack (yığıt) ile iteratif DFS — özyineleme limitinden kaçınmak için."""
    visited = set()
    stack = [start]
    order = []
    while stack:
        node = stack.pop()                   # LIFO — en son gireni çek
        if node not in visited:
            visited.add(node)
            order.append(node)
            # Alfabetik sıra için ters sırada eklenmeli (stack LIFO olduğundan)
            for neighbor in sorted(graph[node], reverse=True):
                if neighbor not in visited:
                    stack.append(neighbor)
    return order


print("DFS (özyinelemeli):", dfs(graph, "Ayşe"))
print("DFS (stack ile)   :", dfs_iterative(graph, "Ayşe"))
````

## ⏱️ Karmaşıklık Analizi (Time & Space Complexity)

**V** = düğüm sayısı (vertices), **E** = kenar sayısı (edges).

| Algoritma | Zaman (Time) | Alan (Space) |
|---|---|---|
| BFS | **O(V + E)** | O(V) — queue + visited |
| DFS (recursive) | **O(V + E)** | O(V) — call stack + visited |
| DFS (iterative) | **O(V + E)** | O(V) — explicit stack + visited |

**Neden O(V + E)?**  
Her düğümü en fazla 1 kez ziyaret ederiz (V), ve her kenardan en fazla 2 kez geçeriz (E — yönsüz grafta). Toplam: O(V + E). *(Wengrow, s.361)*

**Senaryolar:**
- **En iyi (best):** Hedef başlangıcın hemen yanında → ~O(1)
- **Ortalama (average):** Grafın bir kısmı taranır
- **En kötü (worst):** Hedef yok / en sondaysa → tüm graf taranır → O(V + E)

### 🆚 BFS vs DFS — Hangisini Seçmeli?

| Durum | Tercih | Neden |
|---|---|---|
| **En kısa yol** (ağırlıksız graf) | **BFS** | Yakını önce tarar |
| **1. derece bağlantılar** (sosyal ağ) | **BFS** | Hemen komşuları görür |
| **Derinlemesine arama** (oyun ağacı, labirent) | **DFS** | Bir yola hızla iner |
| **Topolojik sıralama, döngü tespiti** | **DFS** | Geri-dönüş kenarlarını bulur |
| **Sınırsız / çok derin graf** | **BFS** | DFS sonsuza inebilir |
| **Bellek kısıtı (geniş graf)** | **DFS** | BFS kuyruğu çok büyüyebilir |


## 🌍 Gerçek Dünya Kullanımları (Real-World Domains)

1. **Sosyal Ağlar (LinkedIn, Facebook):** "X. derece bağlantı" hesabı → **BFS**. Örn. "Bana Leyla kaç adım uzakta?" sorusu BFS ile cevaplanır. *(Wengrow, s.366)*
2. **Web Tarayıcıları & Arama Motorları:** Web sayfaları büyük bir graftır. Googlebot, sayfaları **BFS** (geniş kapsama) veya **DFS** (derin tarama) karışımıyla gezer. Tarayıcının "geri" tuşu → **stack** (DFS mantığı).
3. **Haritalar / Navigasyon (GPS):** Şehirler düğüm, yollar kenar. "Kaç durakta varırım?" → BFS; "En kısa süre / mesafe?" → BFS'in akrabası **Dijkstra** algoritması.
4. **Dosya Sistemleri / Derleyiciler:** Klasör ağacında dosya arama → **DFS** (özyinelemeli klasör tarama). *(Wengrow, s.156)*

## ⚠️ Yaygın Hatalar (Common Pitfalls)

- ❌ **`visited` setini unutmak:** Graf döngüsüyse program **sonsuz döngüye** girer. (Ağaçta gerekmez çünkü döngü yoktur.)
- ❌ **`list.pop(0)` kullanmak:** O(N)! `collections.deque.popleft()` kullan → O(1).
- ❌ **Çok derin grafta özyinelemeli DFS:** Python varsayılan recursion limit ~1000. Çözüm: stack'li (iteratif) DFS.
- ❌ **BFS'te `visited`'e çekerken eklemek:** Aynı düğüm kuyruğa birden çok kez girer. Doğru: kuyruğa **eklerken** visited'e koy.
- ❌ **"Tüm graflar ağaçtır" sanmak:** Tersi geçerli; graf → ağaç olmak için döngüsüz + bağlı + tek kök gerekir.

## 📜 Tarihçe — Kim Buldu? (History)

- **Graf teorisi:** **Leonhard Euler**, **1736** — *Königsberg'in Yedi Köprüsü* problemi. Modern graf teorisinin doğumu.
- **BFS:** İlk bilinen formülasyon **Konrad Zuse**'ye ait (1945, yayımlanmamış doktora tezi). Bağımsız olarak **E. F. Moore** (1959, labirentlerde en kısa yol) ve **C. Y. Lee** (1961, baskılı devre yönlendirmesi) tarafından da türetildi.
- **DFS:** 19. yüzyılda **Charles Pierre Trémaux** tarafından labirent çözümü için kullanıldı. Modern bilgisayar biliminde **Robert Tarjan** ve **John Hopcroft** (1970'ler) DFS'i graf algoritmalarının temeline yerleştirdi.

### 3 İlginç Bilgi (Fun Facts)
1. **Euler aslında çözmedi.** Königsberg probleminde "böyle bir tur **yoktur**" dedi. Bu negatif sonuç koca bir matematik dalı doğurdu.
2. **BFS labirent için doğdu.** Moore'un 1959 makalesinin başlığı: *"The shortest path through a maze."*
3. **Tarjan, DFS ile Turing Ödülü kazandı.** 1986'da Hopcroft ile birlikte; DFS tabanlı *strongly connected components* algoritmasıyla.
