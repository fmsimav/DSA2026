## Ortak Yardımcılar
Aşağıdaki yardımcı fonksiyonlar, algoritmaları test etmek ve doğrulamak için kullanılacaktır.

```python
import random

def is_sorted(a):
    return all(a[i] <= a[i+1] for i in range(len(a)-1))

def demo_list(n=10, seed=42, low=0, high=99):
    random.seed(seed)
    return [random.randint(low, high) for _ in range(n)]

arr = demo_list()
arr

```

## 1) Insertion Sort (Simple Sort)

### Fikir
Sol tarafta büyüyen "sıralı" bir kısım tut ve yeni elemanı bu kısım içinde uygun yerine **içeri sok**. Tıpkı iskambil kağıtlarını elde sıraya dizer gibi.

### Adımlar (küçük örnek)
Dizi: `[5, 2, 4, 6, 1, 3]`
- 2'yi al, [5] içinde doğru yer: `[2, 5]`
- 4'ü al, [2,5] içinde doğru yer: `[2,4,5]`
- 6'yı al, zaten sonda: `[2,4,5,6]`
- 1'i al, başa kaydır: `[1,2,4,5,6]`
- 3'ü al, 2 ile 4 arası: `[1,2,3,4,5,6]`

### Özellikler
- **En iyi**: O(n) (dizi neredeyse sıralıysa)
- **Ortalama / En kötü**: O(n²)
- **Uzay**: O(1) ek alan (in-place)
- **Stability**: **Stable** (eşit elemanların göreli sırası korunur)


```python
def insertion_sort(a):
    a = list(a)  # orijinali bozmayalım

    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        # key'den büyükleri sağa kaydır
        while j >= 0 and a[j] > key:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key
    return a

insertion_sort(arr), is_sorted(insertion_sort(arr))
```

## 2) Bubble Sort

### Fikir
Yanyana elemanları karşılaştır ve ters sıradaysa **yer değiştir**. Her turda **en büyük** eleman sağa "kabarcık gibi" taşınır.

### Özellikler
- **En iyi**: O(n) (hiç takas yoksa erken çıkışla)
- **Ortalama / En kötü**: O(n²)
- **Uzay**: O(1) (in-place)
- **Stability**: **Stable**


```python
def bubble_sort(a):
    a = list(a)
    n = len(a)
    for end in range(n-1, 0, -1):
        swapped = False
        for i in range(end):
            if a[i] > a[i+1]:
                a[i], a[i+1] = a[i+1], a[i]
                swapped = True
        if not swapped:  # erken çıkış
            break
    return a

bubble_sort(arr), is_sorted(bubble_sort(arr))
```

---
## 3) Selection Sort

### Fikir
Her turda **kalan kısmın minimumunu** bul ve baştakiyle **yer değiştir**.

### Özellikler
- **En iyi / Ortalama / En kötü**: O(n²) (karşılaştırma sayısı sabit düzen)
- **Uzay**: O(1) (in-place)
- **Stability**: **Stable değil** (eşit elemanların sırası bozulabilir)


```python
def selection_sort(a):
    a = list(a)
    n = len(a)
    for i in range(n-1):
        min_idx = i
        for j in range(i+1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        if min_idx != i:
            a[i], a[min_idx] = a[min_idx], a[i]
    return a

selection_sort(arr), is_sorted(selection_sort(arr))
```

## 4) Merge Sort

### Fikir

- Diziyi ikiye **böl**, alt dizileri ayrı ayrı **sırala** (özyineleme).
- Sonra iki **sıralı** listeyi **birleştir** (merge).

### Özellikler
- **En iyi / Ortalama / En kötü**: O(n log n)
- **Uzay**: O(n) ek alan (in-place **değil**)
- **Stability**: **Stable** (klasik birleştime ile)


```python
def merge(left, right):
    i = j = 0
    out = []
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            out.append(left[i]); i += 1
        else:
            out.append(right[j]); j += 1
    out.extend(left[i:])
    out.extend(right[j:])
    return out

def merge_sort(a):
    n = len(a)
    if n <= 1:
        return list(a)
    mid = n // 2
    left = merge_sort(a[:mid])
    right = merge_sort(a[mid:])
    return merge(left, right)

merge_sort(arr), is_sorted(merge_sort(arr))
```

## Karşılaştırma Özeti

| Algoritma | En iyi | Ortalama | En kötü | Uzay | In-place | Stable |
|---|---|---|---|---|---|---|
| Insertion (Simple) | O(n) | O(n²) | O(n²) | O(1) | Evet | Evet |
| Bubble | O(n) | O(n²) | O(n²) | O(1) | Evet | Evet |
| Selection | O(n²) | O(n²) | O(n²) | O(1) | Evet | Hayır |
| Merge | O(n log n) | O(n log n) | O(n log n) | O(n) | Hayır | Evet |

> Pratik ipucu: Küçük ve neredeyse sıralı listeler için Insertion Sort çok etkilidir; büyük veri ve tutarlı performans gerektiğinde Merge Sort tercih edilir.

## Hızlı Doğrulama ve Basit Performans Deneyi
Aşağıda tüm algoritmaları aynı veri üzerinde çalıştırıp doğruluyoruz; ayrıca kaba bir süre ölçümü yapıyoruz.

```python
import time

funcs = [
    ("Insertion Sort", insertion_sort),
    ("Bubble Sort", bubble_sort),
    ("Selection Sort", selection_sort),
    ("Merge Sort", merge_sort),
]

data = demo_list(n=200, seed=7)

for name, f in funcs:
    start = time.perf_counter()
    out = f(data)
    dur = (time.perf_counter() - start) * 1000
    print(f"{name:15s}: {dur:8.2f} ms | sorted? {is_sorted(out)}")
```
