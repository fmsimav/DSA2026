# Python'da Hash Veri Yapısı (Dictionary ve Set)

Bu notebook, Python'da **hash tablosu** fikrini ve bunun pratikteki karşılıkları olan **dictionary (`dict`)** ve **set** veri yapılarını, temel seviyede ve örneklerle açıklar.

> Bu içerikte **custom hash fonksiyonu yazımı yoktur**. Amaç, günlük Python kullanımında hash tabanlı veri yapılarının nasıl ve neden kullanıldığını kavramaktır.


## 1) Hash tablosu fikri (yüksek seviye)

Hash tablosu, bir anahtarı (key) sayısal bir değere dönüştürerek veriyi hızlıca bulmayı amaçlayan bir veri yapısıdır.

- Amaç: hızlı **arama / ekleme / silme**
- Günlük örnek: sözlükte kelime → anlam

Python'da bu fikir şu yapılarda karşımıza çıkar:
- `dict` → anahtar–değer
- `set` → benzersiz öğeler


## 2) Dictionary (`dict`)

Dictionary, anahtarları kullanarak değerlere hızlı erişim sağlar.

- Anahtarlar **hashable** (değişmez) olmalıdır
- Değerler her türden olabilir


```python
phonebook = {
    "Ayşe": "0532 000 00 00",
    "Mehmet": "0505 111 11 11",
    "Zeynep": "0544 222 22 22",
}

phonebook["Ayşe"]
```

```python
phonebook["Can"] = "0555 333 33 33"   # ekleme
phonebook["Ayşe"] = "0532 999 99 99" # güncelleme
phonebook
```

```python
phonebook.get("Ali", "(bulunamadı)")
```

### Hashable (anahtar olabilir) tipler

- ✅ `int`, `str`, `float`, `bool`, `tuple`
- ❌ `list`, `dict`, `set`

Anahtarların değişmez olması gerekir.


```python
print(hash(10))
print(hash("abc"))
print(hash((1, 2)))

try:
    hash([1, 2])
except TypeError as e:
    e
```

## 3) Set (`set`)

Set, benzersiz elemanlar tutar ve üyelik testi için çok etkilidir.


```python
nums = [1, 2, 2, 3, 3, 3, 10]
set(nums)
```

```python
# üyelik testi
x = 10
x in nums, x in set(nums)
```

## 4) Zaman karmaşıklığı (ortalama durum)

- `key in dict` → O(1)
- `x in set` → O(1)
- `x in list` → O(n)

Bu nedenle sık arama yapılan durumlarda `set` ve `dict` tercih edilir.


## 5) Mini uygulamalar

### 5.1) Kelime frekansı


```python
text = "veri yapıları ve algoritmalar veri yapıları çok önemli"
words = text.split()

freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1

freq
```

### 5.2) Sıra korunarak tekrar silme


```python
items = ["a", "b", "a", "c", "b", "d"]
seen = set()
unique_items = []

for x in items:
    if x not in seen:
        unique_items.append(x)
        seen.add(x)

unique_items
```

### 5.3) İki listenin kesişimi


```python
a = [1, 2, 3, 4, 5]
b = [4, 5, 6, 7]

set(a) & set(b)
```

## 6) Alıştırmalar ve çözümleri

### Alıştırma 1
Bir listedeki **en sık görülen** elemanı bulun.


```python
lst = [1, 2, 2, 3, 3, 3, 4]

freq = {}
for x in lst:
    freq[x] = freq.get(x, 0) + 1

max(freq, key=freq.get)
```

### Alıştırma 2
İki metindeki ortak kelimeleri bulun.


```python
text1 = "python veri yapıları algoritmalar"
text2 = "veri yapıları çok önemli"

common_words = set(text1.split()) & set(text2.split())
common_words
```
