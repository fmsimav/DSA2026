## 4. Fonksiyonlar

```python
def faktoriyel(n, sep='\n'):
    if n == 0:
        return "safsdf"
    return n * faktoriyel(n-1)

print(faktoriyel(0))
```

## 9. Basit Algoritma Örneği – En Büyük Sayıyı Bulma

```python
def en_buyuk(liste):
    maksimum = liste[0]
    for eleman in liste:
        if eleman > maksimum:
            maksimum = eleman
    return maksimum

print(en_buyuk([3, 7, 2, 9, 5]))
```

## 10. Zaman Karmaşıklığı (Big-O) Notu
- Liste içinde arama: O(n)
- Index ile erişim: O(1)
- Dictionary erişimi: Ortalama O(1)
- İç içe döngüler: Genelde O(n²)
