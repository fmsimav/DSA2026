## 1. Algoritma Nedir?

**Algoritma**, belirli bir problemi çözmek için tanımlı, sonlu sayıda adım içeren, iyi tanımlanmış komutlar dizisidir.

Bir algoritmanın tipik özellikleri:
- **Girdi (input)**: Sıfır veya daha fazla veri alır.
- **Çıktı (output)**: En az bir çıktı üretir.
- **Açıklık (definiteness)**: Her adım açık ve nettir, yorum gerektirmez.
- **Sonluluk (finiteness)**: Sınırlı sayıda adımda biter.
- **Etkinlik (effectiveness)**: Her adım gerçekte uygulanabilir, basit işlemlerden oluşur.

Bu kavramları basit bir örnek üzerinden görelim: "Bir dizideki en büyük sayıyı bul" algoritması.


### 1.1. Örnek: Dizideki En Büyük Elemanı Bulma Algoritması

Sözde kod (pseudocode):

1. `max` değişkenini dizinin ilk elemanına ata.
2. Dizinin geri kalan elemanlarını sırayla dolaş.
3. Eğer mevcut eleman `max`'ten büyükse, `max = mevcut_eleman` yap.
4. Döngü bittiğinde `max` değeri dizideki en büyük elemandır.


```python
numbers = [900, 7, 2, 9, 100, 4, 10, 1]

max_val = numbers[0]  # 1. adım

for num in numbers[1:]:  # 2. adım: diziyi dolaş / slicer
    if num > max_val:    # 3. adım: karşılaştırma
        max_val = num

print("Dizideki en büyük eleman:", max_val)
```

### 1.2. Algoritma Tasarlarken Dikkat Edilecek Noktalar

- Problemi doğru anlamak ve girdileri/çıktıları netleştirmek.
- Kenar durumları (*edge cases*) düşünmek (boş liste, tek elemanlı liste, negatif sayılar vb.).
- Çözümün **doğruluğu** kadar **verimliliğini (efficiency)** de değerlendirmek.

Verimlilik değerlendirmesi için **Big-O notation** kullanacağız.


## 2. Big-O Notation Nedir?

**Big-O notation**, bir algoritmanın **zaman** veya **hafıza** kullanımının, girdi boyutuna (`n`) göre nasıl büyüdüğünü yaklaşık olarak ifade eden bir gösterimdir.

Amaç:
- Detaylardan (donanım, dil, sabit katsayılar) bağımsız olarak, algoritmaların **ölçeklenebilirliğini** kıyaslamak.

Genel büyüme sınıfları:
- **O(1)** – Sabit zaman
- **O(log n)** – Logaritmik zaman
- **O(n)** – Doğrusal zaman
- **O(n log n)** – "n log n" zaman
- **O(n²)** – Kare zaman
- **O(2^n)** – Üstel zaman

Derste yukarıdaki Big-O kaynaklarını grafik ve görseller için kullanabilirsiniz.


### 2.1. O(1) – Sabit Zaman Örneği

Liste içinden **indeks ile** eleman okumak genellikle **O(1)** kabul edilir; çünkü listenin boyutundan bağımsız olarak tek bir erişim işlemi yapılır.


```python
numbers = [10, 20, 30, 40, 50]

# İkinci elemanı (indeks 1) oku – O(1)
print(numbers[1])
```

### 2.2. O(n) – Doğrusal Zaman Örneği

Bir listedeki tüm elemanların toplamını almak için her elemanı **en az bir kez** ziyaret etmek gerekir. Bu yüzden zaman karmaşıklığı **O(n)**'dir.


```python
def sum_list(nums):
    total = 0
    for num in nums:  # n eleman -> n adım
        total += num
    return total

nums = list(range(1, 11))  # 1'den 10'a kadar
print("Liste:", nums)
print("Toplam:", sum_list(nums))
```

### 2.3. O(n²) – Kare Zaman Örneği

İki iç içe döngünün olduğu pek çok algoritma **O(n²)** zaman karmaşıklığına sahiptir. Örneğin, bir listedeki **tüm çiftleri** yazdırmak.


```python
def print_pairs(nums):
    for i in range(len(nums)):         # n adım
        for j in range(len(nums)):     # her adımda n adım daha
            print(f"({nums[i]}, {nums[j]})")

small_list = [1, 2, 3, 4,5,6]
print_pairs(small_list)
```

## 3. Aynı Problemi Farklı Algoritmalarla Çözmek

Örnek problem: **Bir listede eleman arama** ("search")

İki farklı yaklaşım:
1. Doğrusal arama (linear search) – **O(n)**
2. İkili arama (binary search) – **O(log n)**, fakat **sıralı liste** gerektirir.


### 3.1. Doğrusal Arama – O(n)

Her elemanı sırayla kontrol ederiz; elemanı bulana kadar veya listenin sonuna gelene kadar devam ederiz.


```python
def linear_search(nums, target):
    for i, num in enumerate(nums):  # En kötü durumda tüm listeyi gezer -> O(n)
        if num == target:
            return i
    return -1

nums = [5, 3, 8, 4, 2, 9]

print("Liste:", nums)
print("Hedef 4, indeks:", linear_search(nums, 4))
print("Hedef 7, indeks:", linear_search(nums, 7))
```

### 3.2. İkili Arama -Binary Search– O(log n)

İkili arama, **sıralı** bir listede çalışır:
1. Listenin ortasındaki elemanı seç.
2. Hedef ortadakinden küçükse **sol yarıya**, büyükse **sağ yarıya** git.
3. Alt liste üzerinde aynı işlemi tekrarla.

Her adımda listeyi yaklaşık olarak **yarıya böleriz** → karmaşıklık **O(log n)**.


```python
def binary_search(sorted_nums, target):
    left, right = 0, len(sorted_nums) - 1

    while left <= right:
        mid = (left + right) // 2
        mid_val = sorted_nums[mid]

        if mid_val == target:
            return mid
        elif mid_val < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

sorted_nums = [2, 4, 6, 8, 10, 12, 14]

print("Sıralı liste:", sorted_nums)
print("Hedef 10, indeks:", binary_search(sorted_nums, 10))
print("Hedef 7, indeks:", binary_search(sorted_nums, 7))
```

## 4. Basit Bir Deney: n Büyüdükçe Süreler Nasıl Değişiyor?

Aşağıdaki örnek, çok kaba da olsa, farklı algoritmaların çalışma süresini `time` modülü ile ölçmeye yarar. **Gerçek bir performans analizi değildir**, sadece derste fikir vermek için kullanılabilir.


```python
import time

# Büyük bir liste oluşturalım
n = 100_000
large_list = list(range(n))

# Linear search için hedef: sonda olan eleman
start = time.time()
linear_search(large_list, n - 1)
end = time.time()
print(f"Doğrusal arama süresi (n={n}): {end - start:.6f} sn")

# Binary search için aynı hedef (liste zaten sıralı)
start = time.time()
binary_search(large_list, n - 1)
end = time.time()
print(f"İkili arama süresi   (n={n}): {end - start:.6f} sn")
```
